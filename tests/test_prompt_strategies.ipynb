```python
from transformers import AutoTokenizer


tokenizer = AutoTokenizer.from_pretrained('/models/Meta-Llama-3-8B-Instruct', use_fast=False)
```

```python
class DictAttribute:
	def __init__(self, data):
		self.data = data

	def __getattr__(self, name):
		if name in self.data:
			return self.data[name]

	def __contains__(self, key):
		return key in self.data

config = DictAttribute(dict(
	path='/models/datasets/ShareGPT/',
	type='sharegpt',
	data_files='ShareGPT_V4.3_unfiltered_cleaned_split.json',
))
config
```

```python
"conversation" in config
```

```python
from axolotl.prompt_strategies import load


class cfg:
	train_on_inputs = True
	sequence_len = 512

strategy = load(config.type, tokenizer, cfg, config)
strategy
```

```python
import datasets
import os


data = datasets.load_dataset("json", data_files=os.path.join(config.path, config.data_files), split='train')
data
```

```python
data1 = data.select(range(10)).map(strategy.tokenize_prompt)
data1
```
