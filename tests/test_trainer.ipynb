```python
import os


os.environ['https_proxy'] = 'http://172.17.0.1:1081'
os.environ['CUDA_VISIBLE_DEVICES'] = '1'
```

```python
from axolotl.cli import load_cfg, load_datasets, TrainerCliArgs


cfg = load_cfg('../examples/mlpspec/llama3-8b-instruct.yaml')
dataset_meta = load_datasets(cfg=cfg, cli_args=TrainerCliArgs())
dataset_meta
```

```python
dataset_meta.train_dataset[0]
```

```python
from axolotl.utils.models import load_model, load_tokenizer


tokenizer = load_tokenizer(cfg)
tokenizer
```

```python
model, peft_config = load_model(cfg, tokenizer)
model
```

```python
from axolotl.utils.trainer import setup_trainer


trainer = setup_trainer(
    cfg,
    dataset_meta.train_dataset,
    dataset_meta.eval_dataset,
    (model, None, peft_config),
    tokenizer,
    1,
)
trainer
```

```python
trainer.train(resume_from_checkpoint=False)
```
